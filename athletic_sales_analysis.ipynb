```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Define the file paths using Path
sales_2020 = Path('Resources/athletic_sales_2020.csv')
sales_2021 = Path('Resources/athletic_sales_2021.csv')

# Read the CSV files into DataFrames
sales_2020_df = pd.read_csv(sales_2020)
sales_2021_df = pd.read_csv(sales_2021)
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df]).reset_index(drop=True)

combined_sales_df.head()
```

```python
# Check if any values are null.
combined_sales_df.isnull().sum()
```

```python
# Check the data type of each column
combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'])

#check if the data type has been converted
combined_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = combined_sales_df.groupby(
                                                ['region', 'state', 'city']
                                                ).agg(
                                                    {'product': 'count'}
                                                    ).rename(columns={'product': 'Total_Products_Sold'})

# Sort the results in descending order to show the top five regions
total_products_sold = total_products_sold.sort_values(by='Total_Products_Sold', ascending=False)
total_products_sold
```

#### Using `pivot_table`

```python
# Use pivot_table to create a multi-index DataFrame

# and sort the results in descending order to show the top five regions

total_products_sold = combined_sales_df.pivot_table(
    index=['region', 'state', 'city'],
    values='product',
    aggfunc='count'
).rename(columns={'product': 'Total_Products_Sold'}).sort_values(by='Total_Products_Sold', ascending=False) 

# Display the results
total_products_sold.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

total_sales = combined_sales_df.groupby(
                                        ['region', 'state', 'city']
                                        ).agg(
                                         {'total_sales': 'sum'}
                                        ).rename(columns={'total_sales': 'Total Sales'})

# Sort the results in descending order to show the top five regions
total_sales = total_sales.sort_values(by='Total Sales', ascending=False)

# Show the top 5 results
total_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot = combined_sales_df.pivot_table(
    index=['region', 'state', 'city'],
    values='total_sales',
    aggfunc='sum'
).rename(columns={'total_sales': 'Total Sales'})

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.


# Sort the results in descending order to show the top five regions
total_sales_pivot = total_sales_pivot.sort_values(by='Total Sales', ascending=False)

# Display the results
total_sales_pivot.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Group by 'region', 'state', 'city' and aggregate the sum of 'total_sales'
total_sales = combined_sales_df.groupby(
                                        ['region', 'state', 'city']
                                        ).agg(
                                        {'total_sales': 'sum'}
                                        ).rename(columns={'total_sales': 'Total Sales'})

# Sort the results in descending order to show the top five regions
total_sales = total_sales.sort_values(by='Total Sales', ascending=False)

# Show the top 5 results
total_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_pivot = combined_sales_df.pivot_table(
                                                index=['region', 'state', 'city'],
                                                values='total_sales',
                                                aggfunc='sum'
                                                ).rename(columns={'total_sales': 'Total Sales'})

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
total_sales_pivot = total_sales_pivot.sort_values(by='Total Sales', ascending=False)
total_sales_pivot.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_sales = combined_sales_df[combined_sales_df['product'].str.contains("Women's Athletic Footwear")]

womens_sales.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Filter the DataFrame to include only women's athletic footwear sales
womens_athletic_footwear_df = combined_sales_df[combined_sales_df['product'].str.contains("Women's Athletic Footwear")]

# Group by 'retailer', 'region', 'state', 'city' and aggregate the sum of 'units_sold'
units_sold_grouped = womens_athletic_footwear_df.groupby(
                                                        ['retailer', 'region', 'state', 'city']
                                                        )['units_sold'].sum().rename('Womens_Footwear_Units_Sold')

# Convert the result to a DataFrame
units_sold_grouped = units_sold_grouped.reset_index()

# Sort the results in descending order to show the top five retailers
units_sold_grouped = units_sold_grouped.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)

# Show the top 5 results
units_sold_grouped.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"


womens_footwear_units_sold_pivot = womens_athletic_footwear_df.pivot_table(
    index=['retailer', 'region', 'state', 'city'],
    values='units_sold',
    aggfunc='sum'
).rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Sort the results in descending order to show the top five retailers
womens_footwear_units_sold_pivot = womens_footwear_units_sold_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)

# Display the results
womens_footwear_units_sold_pivot.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"

# Create a pivot table with 'invoice_date' as the index and 'total_sales' as the values
sales_pivot_table = combined_sales_df.pivot_table(
                                                index='invoice_date',
                                                values='total_sales',
                                                aggfunc='sum'
                                                ).rename(columns={'total_sales': 'Total Sales'})  # Optionally rename the column

# Display the table
sales_pivot_table.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = sales_pivot_table.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)

sorted_daily_sales.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".

sorted_weekly_sales = womens_athletic_footwear_df.pivot_table(
                                                            index='invoice_date',
                                                            values='total_sales',
                                                            aggfunc='sum'
                                                            ).rename(columns={'total_sales': 'Total Sales'}).resample('W').sum().\
                                                            sort_values(by='Total Sales', ascending=False)

sorted_weekly_sales.head(10)

```
